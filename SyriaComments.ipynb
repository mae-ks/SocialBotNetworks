### There was an out of ordinary comment that appeared the third most frequently. It was about Syria so I thought maybe the creator was trying to raise awareness for the country. This could be a starting point to figuring out how these comments were created.

```python
syriaSameComments = botComments.loc[botComments.message.str.contains("Please take a few minutes to open your eyes, your mind and your heart ") == True]
syriaComments = botComments.loc[botComments.message.str.contains("Syria") == True]
```

```python
print(f"""
Difference: {len(syriaComments)} - {len(syriaSameComments)} = {len(syriaComments) - len(syriaSameComments)}
between General Syria comments and Syria comments with the same message
""")
```

### There is not a sizeable difference in number of comments Syria was mentioned vs the most common comment mentioning Syria. There is still a possibility that the bots commenting the same thing could be commenting something different. I will assume that even if the comment is of a different structure, it will still be about Syria, so we will be using syriaComments as our main database.

```python
syriaSameComments
```

***
### The similarities in their usernames, in that they're mostly using Pascal(camel+snake) case is possibly indicative of a bot network. These bots could perhaps be working towards raising awareness for Syria. I will chack whether all these comments actually do have the same username formats.
***

```python
len(syriaSameComments.username.str.contains('^[A-Z].+_[A-Z].+', regex=True))
```

### Since the length of this regex filtered variable to only check for Pascal Case is equal to that of the syriaSameComments, we can conclude with 100% certainty that these bots all originated from the same network. We can test the same for general comments about Syria.
***

```python
check = syriaComments.username.str.contains('^[A-Z].+_[A-Z].+', regex=True)
syriaComments = syriaComments[check]
syriaComments
```

### There are a total of 1198 unique bot users from these comments, all with the same Pascal name format. I think this is a bot network. I will try to confirm by matching these usernames with the password_ids from our potential bots dataset.

```python
potentialBots = pd.read_csv("C:potential_social_botnets.csv")
potentialBots
```

### Extracting Pascal case usernames from potential bots.

```python
potentialBots = potentialBots[potentialBots.username.str.contains('^[A-Z].+_[A-Z].+', regex=True)]
```

```python
potentialBots.head(10)
```

### Matching syriaComments unique usernames to potentialBots unique usernames to get password_id for each.

```python
allBots = potentialBots[potentialBots.username.isin(set(syriaComments.username.unique()))]
allBots.shape
```

```python
abs(len(allBots.username) - len(syriaComments.username.value_counts()))
```

### There is only a difference of one between all potential bots with Pascal username format and bots that commented about Syria, which means we are only missing one user. I am even more certain now that these bots are working together. I want to compare password_ids now.

```python
allBots.password_id.value_counts()
```
